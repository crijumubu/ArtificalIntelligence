#### INFORMACIÓN GENERAL DE LOS REGISTROS

```python
import wfdb

records = ['16265', '16272', '16273', '16420', '16483', '16539', '16773', '16786', '16795', '17052', '17453', '18177', '18184', '19088', '19090', '19093', '19140', '19830']
signals = []

for i in records:
    
    record = wfdb.rdrecord("Database/{}".format(i))
    signal = record.p_signal
    signals.append(signal)

    print("Registro {}:\n".format(i))
    print("\tLongitud de la señal: {}".format(len(signal)))
    print("\tFrecuencia de la señal: {}\n\n".format(record.fs))
```

```python
from matplotlib import pyplot as plt

signals0 = []
signals1 = []

for signal in signals:
    
    signals0.append(signal[:,0])
    signals1.append(signal[:,1])

print("Dispersión de los datos para signals 0\n")

plot = plt.figure(figsize =(18, 8))
plt.boxplot(signals0)
plt.show()

print("Dispersión de los datos para signals 1\n")

plot = plt.figure(figsize =(18, 8))
plt.boxplot(signals1)
plt.show()
```

#### INFORMACIÓN GENERAL DE LAS ANOTACIONES

```python
notesAll =[]
count = 0

for i in records:
    
    notes = wfdb.rdann("Database/{}".format(i),"atr").sample
    
    print("Anotaciones para el registro {}:\n".format(i))
    print("\tMuestras de las anotaciones: {}".format(notes))
    print("\tTotal de muestras: {}\n\n".format(len(notes)))
    
    notesAll.append(notes)
    count += len(notes)
    
print("Sumatoria de muestras para todos los registros: {}".format(count))
```

#### FORMACIÓN DEL DATASET Y DEL DATAFRAME

```python
import numpy as np
import pandas as pd
from scipy import stats

# Test variable

# total = 0

# Tamano de la ventana
winPartSize = 128 # Tamano medio de la ventana, es decir, que la ventana tendra un tamaño total de 256; para este caso, se hayo en pasos anteriores, que cada segundo se tienen 128 muestras, por lo cual, para este caso, la ventana de 256 representaria una ventana de 2 segundos

# Incremento de la ventana con asolapamiento
increment = winPartSize // 2 # El incremento no sera de media ventana si no de un cuarto de ventana, ya que se hallo que si se tomaba este valor como incremento no se podria tomar en cuenta todas las anotaciones para cada registro (esto unicamente para esta implementacion como tal)

dataSet = []
print('Estado de ejecucion - formacion del dataset:\n')

for i in range (1, len(records)):
    
    notes = notesAll[i]
    signal = signals0[i]
    
    lengthNotes = len(notes)
    lengthSignal = len(signal)
    
    cont = 0
    for j in range(winPartSize, lengthSignal, increment):
        
        beat = 0
        
        # Test outputs
        
        #print('\t\tInferior: ' + str(j - winPartSize))
        #print('\t\tSuperior: ' + str(j + winPartSize))
        #print('\t\tNotes: ' + str(notes[cont]) + '\n')

        if cont < lengthNotes:
            
            if j - winPartSize <= notes[cont] and j + winPartSize >= notes[cont]:
                
                beat = 1
                cont += 1
            
        windows = signal[j - winPartSize: j + winPartSize]
        
        mean = np.mean(windows)
        median = np.median(windows)
        mode = stats.mode(windows)[0]
        standardDeviation = np.var(windows)
        kurtosis = stats.kurtosis(windows, fisher = False)
        
        dataSet.append([mean, median, mode, standardDeviation, kurtosis, beat])

    print('\tRegistro ' + records[i] + ' - check')
    
    # Test outputs
    
    #print('\t\t{}'.format(notes))
    #print('\t\tCont: ' + str(cont) + ', length notes: ' + str(lengthNotes))
    #total += cont

# Test outputs

#print(total)
```

```python
dataframe = pd.DataFrame(data = dataSet)

dataframe.columns = ['Media', 'Mediana', 'Moda', 'Desviación estandar', 'Kurtosis', 'Latido']

dataframe
```

```python
# Toma de las muestras de los latidos del dataframe anterior en un array aparte y eliminacion de la columna, esto con el fin de tener el dataframe con la informacion de las señales y la clasificacion de las mismas por aparte para el entrenamiento de la red neuronal

beats = dataframe['Latido']
beats = beats.to_numpy()

dataframe.pop('Latido')

beats
```

#### CREACIÓN DE LA RED NEURONAL

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(dataframe, beats, train_size = 0.7)
```

#### OPTIMIZACIÓN DE PARAMETROS

```python
max_iter_value = 100
hidden_layer_sizes_value = (100,)
activation_value = "logistic"
learning_rate_init_value = 0.003
```

#### ENTRENAMIENTO DE LA RED NEURONAL

```python
clf = MLPClassifier(
    max_iter = max_iter_value,
    hidden_layer_sizes = hidden_layer_sizes_value,
    activation = activation_value,
    learning_rate_init = learning_rate_init_value
).fit(X_train, y_train)
```

#### RENDIMIENTO DE LA RED NEURONAL

```python
clf.score(X_test, y_test)
```

#### PRUEBA DE LA RED NEURONAL

FORMACIÓN DE LOS DATASETS Y LOS DATAFRAMES DE PRUEBA PARA CADA REGISTRO

```python
winPartSize = 64
rand_position = 278916

test_dataSet = []
test_signals = []
test_beats = []

for i in range(0,len(records)):

    signal = signals0[i][: rand_position]
    lengthSignal = len(signal)
    
    notes = notesAll[i]
    lengthNotes = len(notes)
    
    registry = []
    beats = []
    
    cont = 0
    for j in range(winPartSize, lengthSignal, winPartSize):
        
        beat = 0

        if cont < lengthNotes:
            
            if j - winPartSize <= notes[cont] and j + winPartSize >= notes[cont]:
                
                beat = 1
                cont += 1
        
        windows = signal[j - winPartSize: j + winPartSize]

        mean = np.mean(windows)
        median = np.median(windows)
        mode = stats.mode(windows)[0]
        standardDeviation = np.var(windows)
        kurtosis = stats.kurtosis(windows, fisher = False)

        registry.append([mean, median, mode, standardDeviation, kurtosis])
        beats.append(beat)
        
    test_dataSet.append(registry)
    test_signals.append(signal)
    test_beats.append(beats)
```

```python
print('Dataframe de muestra, este es acorde al registro 16265')

test_dataframe = pd.DataFrame(data = test_dataSet[0])
test_dataframe.columns = ['Media', 'Mediana', 'Moda', 'Desviación estandar', 'Kurtosis']

test_dataframe
```

#### PREDICCIÓN DE LA RED NEURONAL

```python
predictions = []

print('Prediccion para cada regisitro de manera numerica\n')

cont = 0
for dataset in test_dataSet:
    
    test_dataframe = pd.DataFrame(data = dataset)
    test_dataframe.columns = ['Media', 'Mediana', 'Moda', 'Desviación estandar', 'Kurtosis']
    
    predicted = clf.predict(test_dataframe)
    
    print('Prediccion para el registro {}\n'.format(records[cont]))
    print('\t{}\n'.format(predicted[100:200]))
    
    predictions.append(predicted)
    cont += 1
```

```python
print('Prediccion para cada regisitro de manera grafica\n')

x = np.arange(1000)

for i in range(0, len(predictions)):
    
    plt.title('Registro {}'.format(records[i]))
    
    plt.plot(x, predictions[i][:1000])
    plt.show()
```

#### REALIDAD VS PREDICCIÓN

```python
from matplotlib import pyplot

print('Comparacion de la senal real con la prediccion\n')

x = np.arange(1000)

for i in range (0, len(predictions)):
    
    plt.title('Registro {}'.format(records[i]))
    
    plt.plot(x, test_signals[i][:1000])
    plt.plot(x, predictions[i][:1000])
    
    plt.show()
```

#### MATRIZ DE CONFUSIÓN

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

print('Matriz de confusion para cada registro\n')

for i in range(1, len(predictions)):
    
    #print('Registro {}'.format(records[i]))
    
    conf_matrix = confusion_matrix(test_beats[i], predictions[i])
    
    ax = plt.axes()
    sns.heatmap(conf_matrix, ax = ax, annot = True, annot_kws={"size": 16})
    
    ax.set_title('Registro {}'.format(records[i]))
    
    plt.figure()
```
