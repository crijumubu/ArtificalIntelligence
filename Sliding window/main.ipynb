```python
Vobser=[]
signals1=signals[:,0]

for i in anotaciones:
  
  VobserValue = 0

  if i == 649875:
    
    VobserValue = signals1[i-30:]
    
  else:
    
    VobserValue = signals1[i-15:i+15]
    
  Vobser.append(VobserValue)

print("Resultados o salida del modulo:\n\n" + str(Vobser))
```

MODULO 2 DEL PROYECTO

```python
import numpy as np
from matplotlib import pyplot as plt

lengthSignals = len(signals1)

lowerlimit = 0
upperlimit = 15
increment = 7

outputByDeviation = []
outputByMean = []

def insertInOutput (lLimit, uLimit, standardDeviation, mean):
    
    for i in range (lLimit, uLimit):
        
        outputByDeviation.append(standardDeviation)
        outputByMean.append(mean)

while(upperlimit < lengthSignals):
    
    windows = signals1[anotaciones[1] + lowerlimit : anotaciones[1] + upperlimit]
    standardDeviation = np.var(windows)
    mean = np.mean(windows)
    
    insertInOutput(lowerlimit, lowerlimit + increment, standardDeviation, mean)
    
    lowerlimit = lowerlimit + increment
    upperlimit = upperlimit + increment    

insertInOutput(lowerlimit, lengthSignals, 0, 0)  
    
print('Longitud del vector de las muestras: ' + str(lengthSignals))
print('Longitud del vector de resultante por la desviación estandar: ' + str(len(outputByDeviation)))
print('Longitud del vector de resultante por la media: ' + str(len(outputByMean)))

print('\nResultado por la desviación estandar')

plt.plot(outputByDeviation[:1000])
plt.show()

print('Resultado por la media')

plt.plot(outputByMean[:1000])
plt.show()
```
